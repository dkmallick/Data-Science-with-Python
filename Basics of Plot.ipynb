```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
# strip plot
sns.stripplot(y = 'tip',data = tips, jitter = True)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(y = 'tip', x = 'day', data = tips)
plt.ylabel('tip($)')
plt.show()
```

```python
# Swarm plot
sns.swarmplot(y = 'tip', x = 'day', data = tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(y = 'tip', x = 'day', data = tips, hue = 'sex')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(y = 'tip', x = 'day', data = tips, hue = 'sex', orient = 'v')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(y = 'day', x = 'tip', data = tips, hue = 'sex', orient = 'h')
plt.xlabel('tip($)')
plt.show()
```

```python
sns.swarmplot(y = 'day', x = 'tip', data = tips, hue = 'sex')
plt.xlabel('tip($)')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('  ')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgrey')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tips($)')
plt.show()
```

```python
# Joint Plot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
